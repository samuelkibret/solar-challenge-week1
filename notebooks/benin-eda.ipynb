# read csv into data frame

```python
import pandas as pd
df = pd.read_csv("../data/raw/benin-malanville.csv")
df.head()
```

##summary statisitcs for markdown data

```python
df.describe()
```

##Counting missing values and see the columns with over 5% missing data

```python
# Count of missing values per column
missing_values = df.isna().sum()
print (missing_values)

# Filter to see columns with more than 5% missing
threshold = 0.05 * len(df)
missing_values[missing_values > threshold]
```

#Drop columns with over 5% misssing data

```python
df = df.drop(columns=['Comments'])
```

##Checking Z score for outliers and flagging them by creating a column showing rows with z score above 3

```python
from scipy.stats import zscore

# Select relevant columns
cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute Z-scores
z_scores = df[cols_to_check].apply(zscore)

# Find outliers (|z| > 3)
outliers = (z_scores.abs() > 3)

# Count how many outliers per column
outliers.sum()
# check any z -score with above absolute 3
df['Outlier_Flag'] = z_scores.abs().gt(3).any(axis=1)
```

##missing key values

```python
# Check missing values
missing_values = df[cols_to_check].isna().sum()

# Report columns with > 5% missing values
missing_report = (missing_values / len(df)) * 100
missing_over_5 = missing_report[missing_report > 5]

print("Missing value count in key columns:")
print(missing_values)
print("\nColumns with more than 5% missing values:")
print(missing_over_5)
```

#Export cleaned data to data folder and make sure the data is in .gtiignore folder

```python
#export to CSV
df.to_csv("../data/cleaned/benin_clean.csv", index=False)
```

##Timeseries

###start by changing data type to datetime and ensure time sorting

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df = df.sort_values('Timestamp')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
for col in ['GHI', 'DNI', 'DHI']:
    plt.plot(df['Timestamp'], df[col], label=col, alpha=0.7)

plt.title('Solar Irradiance Over Time')
plt.xlabel('Time')
plt.ylabel('Measurement')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Tamb over time

```python
plt.figure(figsize=(14, 6))
plt.plot(df['Timestamp'], df["Tamb"], label="Tamb", alpha=0.7)

plt.title('Tamb Over Time')
plt.xlabel('Time')
plt.ylabel('Celsisus')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['Timestamp'].dt.month
df.groupby('month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean().plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Averages of Solar Irradiance and Temperature')
plt.ylabel('Mean Value')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df['hour'] = df['Timestamp'].dt.hour

df.groupby('hour')[['GHI', 'DNI', 'DHI', 'Tamb']].mean().plot(figsize=(12, 6))
plt.title('Hourly Averages of Solar Irradiance and Temperature')
plt.xlabel('Hour of Day')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Correlation heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_cols = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']
corr = df[corr_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

#Scatterplots

```python
scatter_pairs = [
    ('WS', 'GHI'),
    ('WSgust', 'GHI'),
    ('WD', 'GHI'),
    ('RH', 'GHI'),
    ('RH', 'Tamb')
]

for x, y in scatter_pairs:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=x, y=y, alpha=0.5)
    plt.title(f"{x} vs. {y}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

#Wind rose plot

```python
import numpy as np

# Bin wind directions into 30 intervals
df['WD_bin'] = (df['WD'] // 30) * 30
wind_df = df.groupby('WD_bin')['WS'].mean().reset_index()

# Convert degrees to radians
angles = np.deg2rad(wind_df['WD_bin'])
values = wind_df['WS']

plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
bars = ax.bar(angles, values, width=np.deg2rad(30), color='skyblue', edgecolor='black')

ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_title("Wind Rose (Average WS by WD)")
plt.show()
```
